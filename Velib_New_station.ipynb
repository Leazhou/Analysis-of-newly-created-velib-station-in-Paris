```python
import pandas as pd
import os
pd.options.display.max_columns = None
```

```python
input_folder='DataBrut'
output_folder='DataCleaned'
list_month_num=[4,5,6,7,8]
list_fichier_source=['Velib_April.csv','Velib_May.csv','Velib_June.csv','Velib_July.csv','Velib_Aout.csv']
list_fichier_new_station=['New_station_April.csv','New_station_May.csv','New_station_June.csv','New_station_July.csv','New_station_August.csv']
```

```python
Index_file=0 #0 is for April, 1 is for May, 2 is for June, 3 is for July, 4 is for August
month_num=list_month_num[Index_file]
fichier_source=list_fichier_source[Index_file]
fichier_new_station=list_fichier_new_station[Index_file]
```

```python
df=pd.read_csv(os.path.join(input_folder, fichier_source))
df
```

```python
df.duplicated().sum()
df_cleaned_all=df.drop_duplicates()
```

```python
df_cleaned_all['duedate']=pd.to_datetime(df_cleaned_all['duedate'].str[:-6],format='%Y-%m-%d %H:%M:%S')
```

```python
df_cleaned_all.info()
```

```python
df_cleaned=df_cleaned_all.loc[df_cleaned_all['nom_arrondissement_communes']=='Paris']
```

```python
df_newstation=df_cleaned.loc[(df_cleaned['capacity']==0 )|(df_cleaned['capacity'].isna())|(df_cleaned['is_installed']=='NON')].sort_values(by='duedate')
df_newstation
```

```python
list_code_new_station=df_newstation['stationcode'].unique()
list_newsta_unique=[]
for code_station in list_code_new_station:
    df_one_station=df_cleaned.loc[df_cleaned['stationcode']==code_station].drop_duplicates(subset='stationcode',keep='last')
    list_newsta_unique.append(df_one_station)
df_new_station_unique=pd.concat(list_newsta_unique)
df_new_station_unique=df_new_station_unique[['capacity','name','nom_arrondissement_communes','stationcode','coordonnees_geo']]
df_new_station_unique['month_num']=month_num
df_new_station_unique
```

```python
df_new_station_unique.to_csv(os.path.join(output_folder, fichier_new_station),index=False)
```
